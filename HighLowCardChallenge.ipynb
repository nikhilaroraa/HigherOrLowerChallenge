```python
import random

#function which is used to return a random number between 2-14 representing card numbers
def getCardValue():
    return random.randint(2, 14)

#function used to change card number value into a string based on the corresponding card
def getCardStr(cardValue):
    #if statement for if the card is in between 2 and 9 then the number is returned as a string
    if 2 <= cardValue <= 9:
        return str(cardValue)
    #else statements used to return corresponding letter for numbers between 10-14
    elif cardValue == 10:
        return "T"
    elif cardValue == 11:
        return "J"
    elif cardValue == 12:
        return "Q"
    elif cardValue == 13:
        return "K"
    elif cardValue == 14:
        return "A"
    
#function used to get high or low input from the user while playing the game
def getHLGuess():
    #while loop used to return either "HIGH" or "LOW" depending on the user's string input
    while True:
        guess = input("High or Low (H/L)?: ").upper()
        if guess == "H":
            return "HIGH"
        elif guess == "L":
            return "LOW"

#function used to get the bet amount from the user 
def getBetAmount(maximum):
    while True:
        try:
            #if statement used to check if the bet input is within the amount of points the user has
            bet = int(input(f"Input bet amount (1-{maximum}): "))
            if 1 <= bet <= maximum:
                return bet
        except ValueError:
            print("Invalid input. Please enter a number.")

#function used to check if the player guess was correct or wrong and return the according answer
def playerGuessCorrect(card1, card2, betType):
    #if statement used to check if the HIGH guess is correct
    if betType == "HIGH":
        return card2 > card1
    #else statement used to check if the LOW guess is correct
    elif betType == "LOW":
        return card2 < card1
    
def main():
    #print statements stating instructions of the game
    print("--- Welcome to High-Low ---")
    print("Start with 100 points. Each round a card will be drawn and shown.")
    print("Select whether you think the 2nd card will be Higher or Lower than the 1st card.")
    print("Then enter the amount you want to bet.")
    print("If you are right, you win the amount you bet, otherwise you lose.")
    print("Try to make it to 500 points within 10 tries.")
    print("-------------------------------------")

    # initialized overallPoints and roundCount variable to be used and store points and round numbers accordingly
    overallPoints = 100
    roundCount = 0
    
    #while loop used to play the game within 10 rounds and until the user either reaches 0 or 500 points
    while roundCount < 10 and overallPoints < 500 and overallPoints > 0:
        roundCount += 1
        print(f"\nOVERALL POINTS: {overallPoints} ROUND {roundCount}/10")

        card1Value = getCardValue()
        card1Str = getCardStr(card1Value)
        print(f"First card is a [{card1Str}]")

        guess = getHLGuess()
        bet = getBetAmount(overallPoints)

        card2Value = getCardValue()
        card2Str = getCardStr(card2Value)
        print(f"Second card is a [{card2Str}]")

        if playerGuessCorrect(card1Value, card2Value, guess):
            overallPoints += bet
            print(f"Card1 [{card1Str}] Card 2 [{card2Str}] - You bet '{guess}' for {bet} - YOU WON")
        else:
            overallPoints -= bet
            print(f"Card1 [{card1Str}] Card 2 [{card2Str}] - You bet '{guess}' for {bet} - YOU LOST")

    print("-------------------------------------")
    if overallPoints >= 500:
        print(f"---------------WIN-------------------\nYOU MADE IT TO *{overallPoints}* POINTS IN {roundCount} ROUNDS!")
    elif overallPoints == 0:
        print(f"---------------LOSE------------------\nYOU COULD NOT WIN AT LEAST A ROUND AFTER {roundCount} ROUNDS!")
    else:
        print(f"---------------LOSE------------------\nYOU COULD WIN ONLY *{roundCount - 1}* ROUND AFTER {roundCount} ROUNDS!")

if __name__ == "__main__":
    main()
```
